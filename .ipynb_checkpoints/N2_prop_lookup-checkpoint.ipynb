**物质性质拟合曲线**

数据来源：NIST http://webbook.nist.gov/cgi/fluid
在0.2Mpa下，液氮的沸点为83.636K

__气相数据__(85K-93K)

C_p拟合 J/mol*K

```python
import numpy as np
import matplotlib.pyplot as plt

x = [85,86,87,88,89,90,91,92,93]        
y = [32.784, 32.602, 32.434,32.279,32.135, 32.001,31.876,31.759,31.649]

cof = np.polyfit(x,y,1)          
                                 
p=np.poly1d(cof)                  

plt.plot(x,y,'o',x,p(x),lw=2)

plt.show()

print(cof)
```
